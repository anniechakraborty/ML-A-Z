# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

## Splitting the dataset into the Training set and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print(X_train)
```

## Training the model on the Training Set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```


## Predicting the Test set results

```python
y_hat = regressor.predict(X_test)
```

## Visualising the Training set results

We plot the predited values against the training set values to visualize the error

```python
# plotting the graph inline
%matplotlib inline

# plotting the points of the training set
plt.scatter(X_train, y_train, color='red')

# plotting the curve
plt.plot(X_train, regressor.predict(X_train), color='blue')

# Adding details
plt.title('Training set analysis')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

We plot the predicted values against the test set values to visualize the error

```python
# plotting the graph inline
%matplotlib inline

# plotting the points of the training set
plt.scatter(X_test, y_test, color='red')

# plotting the curve
plt.plot(X_train, regressor.predict(X_train), color='blue')

# Adding details
plt.title('Test set analysis')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
